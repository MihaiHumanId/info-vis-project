```python
df = pd.read_csv(f'{BASE_PATH}/data/housing.csv')
df
```

```python
df.isna().sum()
```

```python
df.dropna(subset = ['Price'], inplace=True)
df.reset_index(drop=True, inplace=True)
df.isna().sum()
```

```python
# Distance - can be somewhat guessed by suburb
for idx, row in df.iterrows():
  if math.isnan(row['Distance']):
    first_match = df[df.Suburb == row.Suburb].iloc[0]
    df.loc[idx, 'Distance'] = first_match.Distance
```

```python
# Postcode - can be somewhat guessed by suburb
for idx, row in df.iterrows():
  if math.isnan(row['Postcode']):
    first_match = df[df.Suburb == row.Suburb].iloc[0]
    df.loc[idx, 'Postcode'] = first_match.Postcode
```

```python
# Bedroom - could probably infer from area but has too many missing values
df['Bedroom2'].fillna(value=round(df['Bedroom2'].mean()), inplace=True)
```

```python
# Bathroom - could probably infer from area but has too many missing values
df['Bathroom'].fillna(value=round(df['Bathroom'].mean()), inplace=True)
```

```python
# Car slots - infer from nearest value with same "Type" and "Rooms" count
for idx, row in df.iterrows():
  if math.isnan(row['Car']):
    first_match = df[(df.Rooms == row.Rooms) & (df.Type == row.Type)].iloc[0]
    if math.isnan(first_match.Car):
      df.loc[idx, 'Car'] = round(df['Car'].mean())
    else:
      df.loc[idx, 'Car'] = first_match.Car
```

```python
# Landsize - infer from nearest value with same Type and Rooms count
for idx, row in df.iterrows():
  if math.isnan(row['Landsize']):
    first_match = df[(df.Rooms == row.Rooms) & (df.Type == row.Type)].iloc[0]
    if math.isnan(first_match.Landsize):
      df.loc[idx, 'Landsize'] = round(df['Landsize'].mean())
    else:
      df.loc[idx, 'Landsize'] = first_match.Landsize
```

```python
# BuildingArea - infer from nearest value with same Type and Rooms count
for idx, row in df.iterrows():
  if math.isnan(row['BuildingArea']):
    first_match = df[(df.Rooms == row.Rooms) & (df.Type == row.Type)].iloc[0]
    if math.isnan(first_match.BuildingArea):
      df.loc[idx, 'BuildingArea'] = round(df['BuildingArea'].mean())
    else:
      df.loc[idx, 'BuildingArea'] = first_match.BuildingArea
```

```python
# YearBuilt - no way to infer
df['YearBuilt'].fillna(value=round(df['YearBuilt'].mean()), inplace=True)
```

```python
# Council area - high chance to overlap with suburb
for idx, row in df.iterrows():
  if pd.isnull(row['CouncilArea']):
    first_match = df[df.Suburb == row.Suburb].iloc[0]
    df.loc[idx, 'CouncilArea'] = first_match.CouncilArea
```

```python
# Latitude - mean of other latitudes inside same suburb or region
for idx, row in df.iterrows():
  if math.isnan(row['Lattitude']):
    suburb_lat_mean = df[df.Suburb == row.Suburb]['Lattitude'].mean()
    if math.isnan(suburb_lat_mean):
      df.loc[idx, 'Lattitude'] = df[df.Regionname == row.Regionname]['Lattitude'].mean()
    else:
      df.loc[idx, 'Lattitude'] = suburb_lat_mean
```

```python
# Longitude - mean of other longitudes inside same suburb or region
for idx, row in df.iterrows():
  if math.isnan(row['Longtitude']):
    suburb_lat_mean = df[df.Suburb == row.Suburb]['Longtitude'].mean()
    if math.isnan(suburb_lat_mean):
      df.loc[idx, 'Longtitude'] = df[df.Regionname == row.Regionname]['Longtitude'].mean()
    else:
      df.loc[idx, 'Longtitude'] = suburb_lat_mean
```

```python
# Regionname - can be inferred from suburb
for idx, row in df.iterrows():
  if pd.isnull(row['Regionname']):
    first_match = df[df.Suburb == row.Suburb].iloc[0]
    df.loc[idx, 'Regionname'] = first_match.Regionname
```

```python
# Propertcount - can be inferred from suburb
for idx, row in df.iterrows():
  if pd.isnull(row['Propertycount']):
    first_match = df[df.Suburb == row.Suburb].iloc[0]
    df.loc[idx, 'Propertycount'] = first_match.Propertycount
```
